# Detecting The License Plates of a Car

### 1. Importing Required Libraries

```python
import cv2
from matplotlib import pyplot as plt
```

### 2. Loading the Image by using the imread function of OpenCV

```python
img = cv2.imread('image4.jpg')
```

### 3. Pre-process the image
<ul>
    <li>Creating a function to convert the image using cv2.Color(img, cv2.COLOR_BGR2RGB) function</li>
    <li>Displaying the image</li>
</ul>

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```

### 4. 
<ul>
    <li>Creating a function for a classifier using the HaarCascade_russian_plate_number.xml file</li>
    <li>Detecting the Car License Plates using the Classifier</li>
</ul>

```python
image_path = 'image2.jpg'
cascade_path = 'haarcascade_russian_plate_number.xml'

def detect_license_plates(image_path, cascade_path):
    # Load the Haar Cascade classifier for license plate detection
    plate_cascade = cv2.CascadeClassifier(cascade_path)
    
    # Read the input image
    img = cv2.imread(image_path)
    
    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect license plates in the image
    plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(100, 40))
    
    # Draw rectangles around the detected plates
    for (x, y, w, h) in plates:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3) # You can change the color and thickness
        
    # Display the image with detected license plates
    cv2.imshow('License Plates', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return plates
```

### 5. Displaying the Results

```python
plates = detect_license_plates(image_path, cascade_path)
print("Detected license plates:", plates)
```
